```python
df=df[['id','title','overview','genre']]
df
```

```python
df['tags']=df['overview']+df['genre']
df
```

```python
df2=df.drop(columns=['overview','genre'])
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(df2['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
df2[df2['title']=='The Godfather'].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key= lambda vector:vector[1])
for i in distance[0:5] :
    print(df2.iloc[i[0]].title)
```

```python
def recommand(movie_title, df, similarity):
  index=df2[df2['title']== movie_title].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key= lambda vector:vector[1])
  for i in distance[1:6] :
    print(df2.iloc[i[0]].title)

# Call the function with the movie title, DataFrame, and similarity matrix
recommand("Iron Man", df, similarity)
```
